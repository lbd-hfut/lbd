```python
import os
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
from PIL import Image
import scipy.io
from scipy.io import loadmat
```

### Noise3

```python
N = 3
u_min = -1; u_max = 1;
Without_cp = loadmat(f'./without cp/Star_N{N}.mat')
plt.figure(figsize=(8,3), dpi=80)
normu = matplotlib.colors.Normalize(vmin=u_min, vmax=u_max)
plt.imshow(Without_cp['v'], cmap='jet', interpolation='nearest', norm=normu)
plt.title("Train: v predicted without cp", fontsize=12)
```

### Noise4

```python
N = 4
u_min = -1; u_max = 1;
Without_cp = loadmat(f'./without cp/Star_N{N}.mat')
R16C16 = loadmat(f'./with cp/noise{N}/Star_N{N}_R16C16.mat')
R32C32 = loadmat(f'./with cp/noise{N}/Star_N{N}_R32C32.mat')
R64C64 = loadmat(f'./with cp/noise{N}/Star_N{N}_R64C64.mat')
plt.figure(figsize=(8,3), dpi=200)
normu = matplotlib.colors.Normalize(vmin=u_min, vmax=u_max)
plt.subplot(2, 2, 1)
plt.imshow(Without_cp['v'], cmap='jet', interpolation='nearest', norm=normu)
plt.title("Train: v predicted without cp", fontsize=8)
plt.axis('off')
plt.subplot(2, 2, 2)
plt.imshow(R16C16['v'], cmap='jet', interpolation='nearest', norm=normu)
plt.title("Train: v predicted with cp R16C16", fontsize=8)
plt.axis('off')
plt.subplot(2, 2, 3)
plt.imshow(R32C32['v'], cmap='jet', interpolation='nearest', norm=normu)
plt.title("Train: v predicted with cp R32C32", fontsize=8)
plt.axis('off')
plt.subplot(2, 2, 4)
plt.imshow(R64C64['v'], cmap='jet', interpolation='nearest', norm=normu)
plt.title("Train: v predicted with cp R64C64", fontsize=8)
plt.axis('off')
```
